```python
class Item:
    def __init__(self, name: str, description: str, damage: int = 0, bonus_attack: int = 0, bonus_defense: int = 0, bonus_magic: int = 0, max_hp: int = 0):
        self.__name = name
        self.__description = description
        self.__bonus_attack = bonus_attack
        self.__bonus_defense = bonus_defense
        self.__bonus_magic = bonus_magic
        self.__max_hp = max_hp
        self.__damage = damage
       # self.__slot = slot
    
    def get_name(self):
        return self.__name
    
    def get_description(self):
        return self.__description
    
    def get_bonus_attack(self):
        return self.__bonus_attack
    
    def get_bonus_defense(self):
        return self.__bonus_defense
    
    def get_bonus_magic(self):
        return self.__bonus_magic
    
    def get_max_hp(self):
        return self.__max_hp
    
    def get_item_details(self):
        item_details = f"{self.get_name()}\n{self.get_description()}\n"
        if self.__damage > 0:
            item_details += f"+{self.__damage} Damage\n"
        if self.__bonus_attack > 0:
            item_details += f"+{self.__bonus_attack} Bonus Attack\n"
        if self.__bonus_defense > 0:
            item_details += f"+{self.__bonus_defense} Defense\n"
        if self.__bonus_magic > 0:
            item_details += f"+{self.__bonus_magic} Magic\n"
        if self.__max_hp > 0:
            item_details += f"+{self.__max_hp} Max HP\n"
        return item_details
    
    def set_name(self, name):
        self.__name = name
    
    def set_description(self, description):
        self.__description = description
    
    def set_bonus_attack(self, bonus_attack):
        self.__bonus_attack = bonus_attack
    
    def set_bonus_defense(self, bonus_defense):
        self.__bonus_defense = bonus_defense
    
    def set_bonus_magic(self, bonus_magic):
        self.__bonus_magic = bonus_magic
    
    def set_max_hp(self, max_hp):
        self.__max_hp = max_hp


class Weapon(Item):
    def __init__(self, name: str, description: str, slot: str, damage: int = 0, bonus_attack: int = 0, bonus_defense: int = 0, bonus_magic: int = 0, max_hp: int = 0):
        super().__init__(name, description, damage=damage, bonus_attack=bonus_attack, bonus_defense=bonus_defense, bonus_magic=bonus_magic, max_hp=max_hp)
        self.__slot = slot
        self.__damage = damage
    def get_damage(self):
        return self.__damage
    def get_slot(self):
        return self.__slot
    
    
    
class Armor(Item):
    def __init__(self, name: str, description: str, slot: str, damage: int = 0, bonus_attack: int = 0, bonus_defense: int = 0, bonus_magic: int = 0, max_hp: int = 0):
        super().__init__(name, description, damage=damage, bonus_attack=bonus_attack, bonus_defense=bonus_defense, bonus_magic=bonus_magic, max_hp=max_hp)
        self.__slot = slot
        self.__damage = damage #this is unnecessary on armor but I am low on time sorry, maybe special armor can increase damage
    def get_slot(self):
        return self.__slot
    def get_damage(self):
        return self.__damage
```
