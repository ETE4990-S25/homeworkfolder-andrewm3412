```python
import import_ipynb
from player_classes import *
from items_classes import *
from inventory_list import *
haschosen = False
stillplaying = True
print("Thank you for choosing to play The Super Awesome Generic Adventure Game Beta!\n")
hasplayed = input("Have you played before? (Y/N)")
if hasplayed.upper() == 'Y':
    reload = input("Would you like to reload? (Y/N)")
    if reload.upper() == 'Y':
        print("Okay, I will load in your old character")
        fileload = 'loadchar.json'
        with open(fileload) as w:
            player_dict = json.load(w)
        p1name = player_dict["name"]
        p1class = player_dict["class_type"]
        if p1class == "Warrior":
            player1 = Warrior(p1name)
            player1.set_max_hp(player_dict["max_hp"])
            player1.set_attack(player_dict["attack"])
            player1.set_defense(player_dict["defense"])
            player1.set_magic(player_dict["magic"])
            haschosen = True
        elif p1class == "Rogue":
            player1 = Rogue(p1name)
            player1.set_max_hp(player_dict["max_hp"])
            player1.set_attack(player_dict["attack"])
            player1.set_defense(player_dict["defense"])
            player1.set_magic(player_dict["magic"])
            haschosen = True
        elif p1class == "Mage":
            player1 = Mage(p1name)
            player1.set_max_hp(player_dict["max_hp"])
            player1.set_attack(player_dict["attack"])
            player1.set_defense(player_dict["defense"])
            player1.set_magic(player_dict["magic"])
            haschosen = True
        elif p1class == "Cleric":
            player1 = Cleric(p1name)
            player1.set_max_hp(player_dict["max_hp"])
            player1.set_attack(player_dict["attack"])
            player1.set_defense(player_dict["defense"])
            player1.set_magic(player_dict["magic"])
            haschosen = True
        elif p1class == "Paladin":
            player1 = Paladin(p1name)
            player1.set_max_hp(player_dict["max_hp"])
            player1.set_attack(player_dict["attack"])
            player1.set_defense(player_dict["defense"])
            player1.set_magic(player_dict["magic"])
            haschosen = True
        else:
            print("An error occured, you're going to have to start a new game")
        invload = 'load_inv_file.json'
        with open(invload) as w:
            itm_dict = json.load(w)
            for items in itm_dict:
                if (items['slot'] == 'weapon') or (items['slot'] == 'offhand'):
                    weapon = Weapon(items["name"], items["description"], items["slot"], items["damage"],
                                items["bonus_attack"], items["bonus_defense"],items["bonus_magic"],
                                items["max_hp"])
                    player1.add_to_inventory(weapon)
                if (items['slot'] == 'head') or (items['slot'] == 'chest') or (items['slot'] == 'legs'):
                    armor = Armor(items["name"], items["description"], items["slot"], items["damage"],
                                items["bonus_attack"], items["bonus_defense"],items["bonus_magic"],
                                items["max_hp"])
                    player1.add_to_inventory(armor)
        print("Player and inventory has been reloaded! You will need to reequip your items")
        
    else:
        print("Thats okay, we will start a new game!")
else:
    print("Thats okay, we will start a new game!")
while haschosen != True:
    player_name = input ("What is your name? ")
    player_type = input("Please choose the role you want to play as \n(Warrior,Rogue, Mage, Cleric, Paladin) ")
    if player_type.title() == "Warrior":
        player1 = Warrior(player_name.title())
        print(f"Congratulations,{player1.get_name()} you are now a {player_type.title()}!")
        print(player1.print_stats())
        haschosen = True
        print("I will load your inventory with some items to mess around with")
        load_inventory(player1)
        
    elif player_type.title() == "Rogue":
        player1 = Rogue(player_name.title())
        print(f"Congratulations,{player1.get_name()} you are now a {player_type.title()}!")
        print(player1.print_stats())
        haschosen = True
        print("I will load your inventory with some items to mess around with")
        load_inventory(player1)
        
    elif player_type.title() == "Mage":
        player1 = Mage(player_name.title())
        print(f"Congratulations,{player1.get_name()} you are now a {player_type.title()}!")
        print(player1.print_stats())
        haschosen = True
        print("I will load your inventory with some items to mess around with")
        load_inventory(player1)
        
    elif player_type.title() == "Cleric":
        player1 = Cleric(player_name.title())
        print(f"Congratulations,{player1.get_name()} you are now a {player_type.title()}!")
        print(player1.print_stats())
        haschosen = True
        print("I will load your inventory with some items to mess around with")
        load_inventory(player1)
        
    elif player_type.title() == "Paladin":
        player1 = Paladin(player_name.title())
        print(f"Congratulations,{player1.get_name()} you are now a {player_type.title()}!")
        print(player1.print_stats())
        haschosen = True
        print("I will load your inventory with some items to mess around with")
        load_inventory(player1) 
        
    else:
        print(f"That is not a type, please try again")
        
print("Here is your inventory:\n")
player1.display_inventory()

while stillplaying:
    choice = None
    item_choice = None
    slot_choice = None
    print("\nWhat would you like to do? Choose an option: ")
    #display menu
    print("1. Display Inventory \n2. Get an item's information \n3. Equip an Item \n4. Unequip an Item\n5. Display"
          + " character stats \n6. Remove an item from inventory \n7. Create and Add an item to inventory \n8. Save and Quit")
    choice = input("Please Choose a number from 1-8: ")
    if int(choice) == 1:
        player1.display_inventory()
    elif int(choice) == 2:
        item_choice = input("\nPlease enter the name of the item you wish to know more about ")
        item_choice = item_choice.title()
        print(player1.item_description(item_choice))
    elif int(choice) == 3:
        item_choice = input("\nPlease enter the name of the item you wish to equip ")
        item_choice = item_choice.title()
        player1.equip(item_choice)
    elif int(choice) == 4:
        slot_choice = input("\nEnter the slot that you wish to unequip an item from ")
        slot_choice = slot_choice.lower()
        player1.unequip(slot_choice)
    elif int(choice) == 5:
        print(player1.print_stats())
    elif int(choice) == 6:
        item_choice = input("\nWhich item would you like to remove ")
        item_choice = item_choice.title()
        player1.remove_from_inventory(item_choice)
    elif int(choice) == 7:
        item_name = None
        item_descrp = None
        health = 0
        attack = 0
        defense = 0
        magic = 0
        damage = 0
        item_choice = input("\nChoose what kind of item: Weapon/Armor ")
        if item_choice.title() == "Armor":
            item_name = input("Please name the armor ")
            item_name = item_name.title()
            item_slot = input("Which slot is it for? (head, chest, legs) ")
            item_slot = item_slot.lower()
            item_descrp = input("Please provide a breif description of your item ")
            health = input("How much bonus health will it provide? ")
            health = int(health)
            attack = input("How much bonus attack will it provide? ")
            attack = int(attack)
            defense = input("How much bonus defense will it provide? ")
            defense = int(defense)
            magic = input("How much bonus magic will it provide? ")
            magic = int(magic)
            newitem = Armor(item_name, item_descrp, item_slot, 0, attack, defense, magic, health)
            player1.add_to_inventory(newitem)
            print("The new armor has been added to your inventory! ")
            
        elif item_choice.title() == "Weapon":
            item_name = input("Please name the weapon ")
            item_name = item_name.title()
            item_slot = input("Which slot is it for? (weapon, offhand) ")
            item_slot = item_slot.lower()
            item_descrp = input("Please provide a breif description of your item ")
            damage = input("How much damage will it do? ")
            damage = int(damage)
            health = input("How much bonus health will it provide? ")
            health = int(health)
            attack = input("How much bonus attack will it provide? ")
            attack = int(attack)
            defense = input("How much bonus defense will it provide? ")
            defense = int(defense)
            magic = input("How much bonus magic will it provide? ")
            magic = int(magic)
            newitem = Weapon(item_name, item_descrp, item_slot, damage, attack, defense, magic, health)
            player1.add_to_inventory(newitem)
            print("The new weapon has been added to your inventory!")
            
        else:
            print("If you can't spell correctly do you really think you should have this kind of power?")
    elif int(choice) == 8:
        print("\nSaving your character and quiting, this will remove all equiped items")
        player1.unequip('weapon')
        player1.unequip('offhand')
        player1.unequip('head')
        player1.unequip('chest')
        player1.unequip('legs')
        save_dict = ({
            "name": player1.get_name(),
            "class_type": player1.get_class_type(),
            "max_hp": player1.get_max_hp(),
            "attack": player1.get_attack(),
            "defense": player1.get_defense(),
            "magic": player1.get_magic()
        })
        loadfile = 'loadchar.json'
        with open(loadfile, 'w') as w:
            json.dump(save_dict, w)
        inventory_dict = []
        for item in player1.get_inventory():
            item_dict = {"name": item.get_name(),
                         "description": item.get_description(),
                         "slot": item.get_slot(),
                         "damage": item.get_damage(),
                         "bonus_attack": item.get_bonus_attack(),
                         "bonus_defense": item.get_bonus_defense(),
                         "bonus_magic": item.get_bonus_magic(),
                         "max_hp": item.get_max_hp()}
            inventory_dict.append(item_dict)
        load_inv_file = 'load_inv_file.json'
        with open(load_inv_file, 'w') as w:
            json.dump(inventory_dict, w)
        print ("Game was saved!")
        break
        
    else:
        continue
print("Thank you for testing out the game!")
```
