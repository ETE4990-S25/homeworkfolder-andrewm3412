```python
class PlayerType(object):
    """Create a model of a player"""
    def __init__(player, name, skill, strength, powers, armor): #innit is to initiate
        """Initialize name & attributes."""
        player.name = name
        player.skill = skill
        player.strength = strength
        player.powers = powers
        player.armor = armor
```

```python
import json

startup = ""
active = True

get_name = input("Tell me your name: ")
print ("Welcome to the fight, " + get_name + " choose your role: ")


active = True
while active:
    startup = input ("Choose you player type: Hero, Villain, or Civilian? = ")
    if startup in ['Hero', 'Villain', 'Civilian']:
        active = False
    else: 
        print(startup)
        
if startup == 'Hero' :
    player = PlayerType(get_name, 5, 7, 8, 5)
elif startup == 'Villain' :
    player = PlayerType(get_name, 8, 5, 7, 6)
elif startup == 'Civilian' :
    player = PlayerType(get_name, 5, 2, 0, 0)
    
print('\nPlayer Name: ' + player.name + '\n' +'\nHere are your stats: '+ '\nSkill: ' + str(player.skill) + '\nStrength: ' + str(player.strength) + '\nPowers: ' + str(player.powers) + '\nArmor: ' + str(player.armor))
print('\nChoose your weapon!')


# read the inventory JSON files for each player type
with open('inventoryhero.json') as f:
    inventory_hero = json.load(f)

with open('inventoryvillain.json') as f:
    inventory_villain = json.load(f)

with open('inventorycivilian.json') as f:
    inventory_civilian = json.load(f)

# display the inventory for the chosen player type
if startup == 'Hero':
    for item in inventory_hero:
        if item['attribute'] == player.skill:
            print(item['item_name'] + ': ' + str(item['quantity']))
            
elif startup == 'Villain':
    for item in inventory_villain:
        if item['attribute'] == player.powers:
            print(item['item_name'] + ': ' + str(item['quantity']))
            
elif startup == 'Civilian':
    for item in inventory_civilian:
        if item['attribute'] == player.armor:
            print(item['item_name'] + ': ' + str(item['quantity']))

# display the inventory for the chosen player type
if startup == 'Hero':
    print('\nInventory for Hero:')
    for item in inventory_hero:
            print(item['item_name'] + ': ' + str(item['attribute']) + ' ' + str(item['description']))
    done = False
    equipped_item = None
    while not done:
        choice = input("\nEnter the name of the weapon you want to Equip/Dequip or Quit: ")
        if choice == 'Quit':
            print('\nGood luck saving the world!')
            done = True
        else:
            for item in inventory_hero:
                if item['item_name'] == choice:
                    if item == equipped_item:
                        print(f"You have unequipped {choice}.")
                        equipped_item = None
                    else:
                        print(f"You have equipped {choice}.")
                        equipped_item = item
                    break
            else:
                print("Invalid choice. Please try again.")

        
elif startup == 'Villain':
    print('\nInventory for Villain:')
    for item in inventory_villain:
            print(item['item_name'] + ': ' + str(item['attribute']) + ' ' + str(item['\ndescription']))
    done = False
    equipped_item = None
    while not done:
        choice = input("\nEnter the name of the weapon you want to Equip/Dequip or Quit: ")
        if choice == 'Quit':
            print('\nGood luck taking over the world!')
            done = True
        else:
            for item in inventory_villain:
                if item['item_name'] == choice:
                    if item == equipped_item:
                        print(f"You have unequipped {choice}.")
                        equipped_item = None
                    else:
                        print(f"You have equipped {choice}.")
                        equipped_item = item
                    break
            else:
                print("Invalid choice. Please try again.")


        
elif startup == 'Civilian':
    print('\nInventory for Civilian:')
    for item in inventory_civilian:
            print(item['item_name'] + ': ' + str(item['attribute']) + ' ' + str(item['description']))
    done = False
    equipped_item = None
    while not done:
        choice = input("\nEnter the name of the weapon you want to Equip/Dequip or Quit: ")
        if choice == 'Quit':
            print('\nWhat are you still doing here? RUN!!!')
            done = True
        else:
            for item in inventory_civilian:
                if item['item_name'] == choice:
                    if item == equipped_item:
                        print(f"You have unequipped {choice}.")
                        equipped_item = None
                    else:
                        print(f"You have equipped {choice}.")
                        equipped_item = item
                    break
            else:
                print("Invalid choice. Please try again.")

else:
    print("Invalid player type.")
```

```python
def save_game(player, startup):
    data = {'name': player.name, 'skill': player.skill, 'strength': player.strength, 'powers': player.powers, 'armor': player.armor, 'type': startup}
    with open('saved_game.json', 'w') as f:
        json.dump(data, f)
# after the player has chosen their weapon and the game is about to end
save_game(player, startup)
print("Game saved. Goodbye!")
def load_game():
    with open('saved_game.json', 'r') as f:
        data = json.load(f)
    return data
```

```python
# load saved game data if available
try:
    data = load_game()
    player = PlayerType(data['name'], data['skill'], data['strength'], data['powers'], data['armor'])
    startup = data['type'], data['skill'], data['strength'], data['powers'], data['armor']
    print("Saved game loaded. Welcome back, " + player.name + "!")
    print("Player type & attributes: ", startup)
except FileNotFoundError:
    print("No saved game data found. Starting new game...")
```
