````python
class Farm:

    def __init__(self, animal, purpose):
        self.animal = animal
        self.purpose = purpose

    def get_purpose(self):
        return f"{self.animal} are useful for {', '.join(self.purpose)}"
    

# Creating animal objects

cow = Farm(animal='cow', purpose=['Milk', 'Meat', 'Leather'])
pigs = Farm(animal='pigs', purpose=['Meat', 'Skin', 'Helps soil fertility'])
dogs = Farm(animal='dogs', purpose=['Herding', 'Protection'])
goats = Farm(animal='goats', purpose=['Milk', 'Meat', 'Brush control'])

animals =[cow, pigs,dogs,goats]
# Printing their purposes
for animal in animals:
    print(animal.get_purpose())

print("[1] Add an animal")
print("[2] remove an animal")

choice= input("\nEnter your choice: ").strip()

if choice == "1":
    add = input("\nEnter a new animal: ").strip().lower()
    if add not in [a.animal for a in animals]:
        purposes = input(f"Enter purposes for {add} (comma-separated): ").split(',')
        animals.append(Farm(add, [p.strip() for p in purposes]))
        print(f"\nAdded {add.capitalize()} to the farm!")
    else:
        print(f"\n{add.capitalize()} is already on the farm.")


# Print updated list

elif choice == "2":
    remove = input("Enter to remove anima: ").strip().lower()
    for a in animals:
        if a.animal == remove:
            animals.remove(a)
            print(f"\n{remove.capitalize()} have been remove from the farm.")
            break
    else:
        print(f"\n{remove.capitalize()} have been not found on the farm.")

print("\nUpdated farm animals:")
for animal in animals:
    print(animal.get_purpose()) 
````
